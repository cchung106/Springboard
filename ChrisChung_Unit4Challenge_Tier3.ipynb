```python
#Creating new column 'Year' and pulling the datetime year from 'Month' column
properties_clean['Year'] = pd.DatetimeIndex(properties_clean['Month']).year
properties_clean.head()
```

```python
#Create new DF properties_grouped, with average yearly prices for each borough and for each year
properties_grouped = properties_clean.groupby(['London_Borough', 'Year'])["Average_Price"].mean()
```

```python
properties_grouped.head()
```

```python
#Reset index of this DF
properties_grouped = properties_grouped.reset_index()
```

```python
properties_grouped.head(10)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(borough_name):
    """Calculate the ratio of house prices, comparing the price of house in 2018 to price in 1998"""
    
    #subset the DF based on borough name
    properties_filtered=properties_grouped[properties_grouped['London_Borough']==borough_name]
    
    #Get Average Price for that borough in 1998
    avg_price_1998 = properties_filtered.loc[properties_filtered['Year']==1998, 'Average_Price'].values[0]

    
    #Get Average Price for that borough in 2018
    avg_price_2018 = properties_filtered.loc[properties_filtered['Year']==2018, 'Average_Price'].values[0]

    
    #Calculate ratio
    ratio = avg_price_2018 / avg_price_1998
    
    return ratio
```

```python
#Testing the function using 'Westminster' as borough name
create_price_ratio("Westminster")
```

```python
#Creating an empty dictionary to store boroughs as keys and price ratios as values
price_ratio_dict = {}
```

```python
#Iterate over unique London Boroughs and compute price ratio, add it to the dictionary.
for borough_name in properties_grouped['London_Borough'].unique():
    price_ratio_dict[borough_name] = create_price_ratio(borough_name)
```

```python
print(price_ratio_dict)
```

```python
#Convert price_ratio_dict into pandas dataframe
price_ratio_df = pd.DataFrame(list(price_ratio_dict.items()),columns = ['London_Borough','Price_Ratio']) 

print(price_ratio_df.head())
```

```python
#Sort price_ratio_df by price_ratio
top15 = price_ratio_df.sort_values('Price_Ratio', ascending = False).head(15)
print(top15)
```

```python
top15.plot(x="London_Borough", y="Price_Ratio", kind='bar', xlabel='London Boroughs', ylabel='Price Ratio', title='Price Ratios of Homes for London Boroughs from 1998 to 2018')
plt.show()
```
